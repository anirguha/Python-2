```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
names = ["Abraham","Guido","Alex","Constance"]
ages= [23, 40, 65,82]
salary = [23.5, 92.9, 21.8, 65.3]
```

```python
df = pd.DataFrame({"age": ages, "salary": salary},index=names)
```

```python
df
```

```python
print(df.loc['Abraham','age'])
```

```python
print(df.iloc[0,0])
```

```python
#Loop in through the columns of the DataFrame
for col in df:
    print(col,df[col])
```

```python
#Loop in through the rows of the DataFrame uding iterrows()

rows = list(df.iterrows())
rows

for key, values in rows:
  print(f'Key:{key}\nValues\n{values}')
```

```python
#Loop in through the rows using itetuple()
for rows in df.itertuples():
    print(f'Key:{rows.Index}\nage:{rows.age}\nsalary:{rows.salary}')
```

```python
#using Index based opersations
for i in range(len(df)):
    print(f'Key:{df.index[i]}\n{df.iloc[i]}')
```

```python
ax = df.salary.hist()
ax.set_xlabel(salary.index)
```
