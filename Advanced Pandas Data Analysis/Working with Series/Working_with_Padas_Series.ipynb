## Data loading

### Subtask:
Load the data from "Iris.csv" into a pandas DataFrame.


**Reasoning**:
Load the data from "Iris.csv" into a pandas DataFrame, display the first 5 rows, and show the column names and their data types to verify the data loading and understand the structure.



```python
import pandas as pd

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('Iris.csv')

# Display the first 5 rows of the DataFrame
display(df.head())

# Display the column names and their data types
display(df.info())
```

## Data exploration

### Subtask:
Explore the dataset to understand its structure, summary statistics, and the distribution of features and the target variable.


**Reasoning**:
Display the shape of the DataFrame, generate descriptive statistics for numerical columns, get the value counts for the 'Species' column, and check for missing values to understand the dataset's structure and distribution.



```python
# Display the shape of the DataFrame
print("Shape of the DataFrame:")
display(df.shape)

# Generate descriptive statistics for numerical columns
print("\nDescriptive statistics for numerical columns:")
display(df.describe())

# Get the value counts for the 'Species' column
print("\nValue counts for the 'Species' column:")
display(df['Species'].value_counts())

# Check for missing values in the DataFrame
print("\nMissing values per column:")
display(df.isnull().sum())
```

## Data visualization

### Subtask:
Create visualizations such as scatter plots to show the relationships between different features and the target variable (species).


**Reasoning**:
Import the necessary libraries for plotting and create a pair plot to visualize the relationships between all pairs of numerical features, colored by Species. This addresses steps 1, 2, 3, and 4 of the instructions and provides a comprehensive overview of the feature relationships.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a pair plot of the numerical features, colored by 'Species'
# Drop the 'Id' column as it is not a numerical feature relevant for the plot
sns.pairplot(df.drop("Id", axis=1), hue="Species", diag_kind="hist")

# Add a title to the plot for clarity (this title will appear above the entire grid)
plt.suptitle("Pair Plot of Iris Dataset Features by Species", y=1.02)

# Display the plot
plt.show()
```
