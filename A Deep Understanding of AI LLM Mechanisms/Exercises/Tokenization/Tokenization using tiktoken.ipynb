```python
import tiktoken
import requests
```

```python
URL = "https://www.gutenberg.org/cache/epub/84/pg84.txt"

text = requests.get(URL).text
```

```python
len(text)
```

```python
tokenizer = tiktoken.get_encoding("o200k_base")
tokens = tokenizer.encode(text)
```

```python
len(tokens)
```

```python
tokenizer.decode(tokens[:10])
```
