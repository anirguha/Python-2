### head() and tail() functions

```python
Cardata = { "Mercedes": [2, 4, 0, 4, 0, 3], "Ford": [3, 0, 0, 1, 6, 12], "Tata":[9, 3, 4, 1, 0, 0], "Renault":[12, 1, 0, 0, 3, 1]}
Carsales = pd.DataFrame(Cardata)
Carsales.index.rename("Sales place", inplace=True)
```

```python
Carsales_N = pd.concat([Carsales]*5000,ignore_index=True)
```

```python
Carsales_N.head()
```

```python
Carsales_N.tail()
```

```python
#Selection from the middle of DataFrame
Carsales_N.iloc[5000:5200,:]
```
