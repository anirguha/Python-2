### Code to Generate Histograms

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore","use_inf_as_na", FutureWarning)
```

```python

# Load iris dataset from seaborn
iris = sns.load_dataset("iris")
print(iris)
```

```python
#Plot Histogram on iris dataframe
ax = iris.plot.hist(column=["sepal_length"],bins=50,grid=True)
ax.set_title("Histogram of sepal length of 150 flowers",fontsize=16)
ax.set_xlabel("Sepal Length in cm",fontsize=12)
ax.set_ylabel("Count",fontsize=12)
plt.show()
```

```python
x = np.random.randn(1000000)
# Plot histogram of random numbers
plt.hist(x, bins=100, edgecolor='black', alpha=0.7)
plt.title("Histogram of Random Numbers", fontsize=16)
plt.xlabel("Value", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
dfx = pd.DataFrame(x, columns=['Random Numbers'])
print("Skewness of the Distribution:",dfx.skew().values[0])  # Accessing the first value of the Series returned by skew()
print("Kurtosis of the Distribution:", dfx.kurtosis().values[0]+3)  # Adding 3 to match the Fisher definition of kurtosis
```

```python
#Draw Histogram for each of the species of flower
from matplotlib.pyplot import subplots


fig, ax = subplots()
colors = {"setosa":"red","versicolor":"blue","virginica":"yellow"}
for Species,data in iris.groupby("species"):
    data.plot(kind="hist",column="sepal_length",bins=15,color=colors[Species],alpha=0.5,ax=ax)
ax.set_title("Histogram of Sepal Length of 3 Species", fontsize=16)
ax.set_xlabel("Sepal Length in cm", fontsize=12)
ax.set_ylabel("Count", fontsize=12)
ax.legend(colors.keys(), title="Species",loc="upper left")
fig.suptitle("")
plt.show()
```

```python
#Plot Histogram with seaborn
sns.histplot(data=iris,x="sepal_length",bins=50)
plt.show()
```

```python
#Plot Histogram with seaborn with hues
sns.histplot(data=iris,x="sepal_length",bins=25,hue="species")
plt.show()
```

```python
#Plot Histogram with seaborn with step function
sns.histplot(data=iris,x="sepal_length",bins=25,hue="species",element="step").set(title="Histogram of Sepal Length with Step Function")
plt.show()
```

```python
#Plot Histogram with seaborn with two categories
sns.histplot(data=iris,x="sepal_length",y="sepal_width",bins=25,hue="species",element="step").set(title="Histogram of Sepal Length for two categories")
plt.show()
```

```python
#4-Histograms-in-1-graph joined
graph = sns.JointGrid()
sns.histplot(data=iris,x="sepal_length",y="sepal_width",bins=25,hue="species",alpha=0.6,ax=graph.ax_joint)

sns.histplot(data=iris,x="sepal_length",bins=25,hue="species",alpha=0.6,ax=graph.ax_marg_x)

sns.histplot(data=iris,y="sepal_width",bins=25,hue="species",alpha=0.6,ax=graph.ax_marg_y)
graph.figure.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9)
graph.figure.suptitle("4 Histograms in 1 Graph for 2 variables in dataset", fontsize=16)
plt.show()
```
