### Duplicate Handling

```python
import pandas as pd
```

```python
Cardata = {"Mercedes": [2,4,0,4,0,3], "Ford":[3,0,0,1,6,2], 
           "Tata":[9,3,4,1,0,0], "Renault":[12,1,0,0,3,1]}
Carsales = pd.DataFrame(Cardata)
Carsales.index.rename("Sales place", inplace=True)
Carsales.rename(index={0:"One",1:"Two",2:"Three",3:"Four",4:"Five",5:"Six"},inplace=True)

Carsales.insert(0, "Sales_place_name", ["Europe 1", "Australia 1", "USA 1", "Asia 1", "Africa 1", 
                                        "South America 1"], allow_duplicates=True)

Newsalesplace2 = pd.DataFrame({"Sales_place_name": "South America 1", "Mercedes":3, "Ford":2,
                               "Tata": 1, "Renault": 1}, index= ["Seven"])
Newsalesplace3 = pd.DataFrame({"Sales_place_name": "Asia 1", "Mercedes":4, "Ford":1,
                               "Tata": 1, "Renault": 0}, index= ["Eight"])

Newsalesplace2
```

```python
Carsales = pd.concat([Carsales, Newsalesplace2, Newsalesplace3])
Carsales
```

```python
#Mark duplciates in the DataFrame based on a key (first occurrence
#is marked as False and subsequent ones as True by default
Carsales.duplicated(subset=["Sales_place_name"])
```

```python
#Mark duplciates in the DataFrame based on a key (first occurrence
#is marked as False and subsequent ones as True by default
Carsales.duplicated(subset=["Sales_place_name"],keep="last")
```

```python
#Mark duplciates in the DataFrame based on all columns(first occurrence
#is marked as False and subsequent ones as True by default
Carsales.duplicated()
```

```python
#Mark duplciates in the DataFrame based on multiple columns(first occurrence
#is marked as False and subsequent ones as True by default
Carsales.duplicated(subset=["Sales_place_name","Mercedes"])
```

```python
#Remoce duplciates in the DataFrame based on a key column(first occurrence
#is marked as False and subsequent ones as True by default
Carsales_rem1=Carsales.drop_duplicates(subset=["Sales_place_name"])
Carsales_rem1
```

```python
#Remoce duplciates in the DataFrame based on all columns(first occurrence
#is marked as False and subsequent ones as True by default
Carsales_rem2=Carsales.drop_duplicates()
Carsales_rem2
#Eight is removed as it is a duplicate of Four
```
