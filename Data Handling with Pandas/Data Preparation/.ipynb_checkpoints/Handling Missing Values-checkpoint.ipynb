### Handling Missing Data and Imputation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
```

```python
Cardata = {"Mercedes": [2,4,None,4,0,3], "Ford":[3,0,0,1,6,2], 
           "Tata":[9,3,4,1,0,0], "Renault":[12,1,None,None,3,1]}
Carsales_mv = pd.DataFrame(Cardata)
Carsales_mv.index.rename("Sales place", inplace=True)
Carsales_mv.rename(index={0:"One",1:"Two",2:"Three",3:"Four",4:"Five",5:"Six"},inplace=True)
Carsales_mv
```

```python
#Descriptive statistics of missing values
Carsales_mv_isnull = Carsales_mv.isnull()
Carsales_mv_isnull
```

```python
#Showing count of missing values
Carsales_mv.isnull().sum()
```

```python
#Descriptive statistics without missing values
Carsales_mv.describe()
```

```python
#Drop rows/ columns containing missing values from DataFrame
Carsales_2 = Carsales_mv.dropna() #drops rows by default
Carsales_3 = Carsales_mv.dropna(axis=1) #drops columns
print(Carsales_2)
print(Carsales_3)
```

```python
#Fill the missing values with zeros
Carsales_zfill = Carsales_mv.fillna(0,inplace=False)
Carsales_zfill
Carsales_mv
```

```python
#Imputation technique mean-filled missing values
Carsales_mv_mean = Carsales_mv.mean()
Carsales_mv_meanfilled = Carsales_mv.fillna(Carsales_mv_mean,inplace=False)
Carsales_mv_meanfilled
```

### Create a DataFrame from Geyser dataset from seaborn moudle

```python
geyser = sns.load_dataset("geyser")
geyser
```

```python
geyser2 = geyser.copy(deep=True)
print(geyser2)
```

```python
#Replace ggeyser2 DataFrame with some missing values at indexes
#specified by nlist1 and nlist2 in columns duration and waiting
nlist1 = [4, 34, 35, 57, 69, 82, 234, 267]
nlist2 = [3, 6, 73, 89, 24]

for n in nlist1:
    geyser2.loc[n,"duration"] = None

for n in nlist2:
    geyser2.loc[n,"waiting"] = None

geyser3 = geyser2.copy(deep=True)
geyser2
```

```python
#Impute the missing values in geyser2 using linear interplolation
#force the object types to float/ int
geyser2["duration"] = pd.to_numeric(geyser2["duration"], errors="coerce")
geyser2["waiting"] = pd.to_numeric(geyser2["waiting"], errors="coerce")

#interpolate the values
geyser2["duration"]=geyser2["duration"].interpolate(method="linear")
geyser2["waiting"]=geyser2["waiting"].interpolate(method="linear")
```

```python
geyser2
```

```python
#plot the DataFrame
fig, ax = plt.subplots()
geyser2.plot.scatter(x="duration",y="waiting",color="Blue",marker="x",ax=ax)
geyser.plot.scatter(x="duration",y="waiting",color="Red",marker="o",ax=ax)
plt.show()
```

```python
#Wide variation between estimated and actual value
#Using linear regression model to estimate the missing values
#Using scikit-learn to create a linear regerssion model on the DataFrame without missing values

geyser4_dmv = geyser3.dropna()
model_1 = linear_model.LinearRegression()
model_2 = linear_model.LinearRegression()
model_1.fit(X=geyser4_dmv[["duration"]],y=geyser4_dmv["waiting"])
model_2.fit(X=geyser4_dmv[["waiting"]],y=geyser4_dmv["duration"])
```

```python
#To test the linear model, the null/ missing values need to be replaced.
#Replacing the missing values with a very high value.
geyser3["duration"] = geyser3["duration"].fillna(10000)
geyser3["waiting"] = geyser3["waiting"].fillna(10000)

geyser3["est_duration"]=model_2.predict(geyser3[["waiting"]])
geyser3["est_waiting"]=model_1.predict(geyser3[["duration"]])

#Replace the extreme values with estimated values
geyser3.loc[geyser3['duration']>5000, 'duration'] = geyser3["est_duration"]
geyser3.loc[geyser3['waiting']>5000, 'waiting'] = geyser3["est_waiting"]
geyser3
```

```python
#plot the DataFrame
fig, ax = plt.subplots()
geyser3.plot.scatter(x="duration",y="waiting",color="Blue",marker="x",ax=ax)
geyser.plot.scatter(x="duration",y="waiting",color="Red",marker="o",ax=ax)
plt.show()
```
