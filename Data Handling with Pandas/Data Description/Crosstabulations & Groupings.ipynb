### Crosstabulations & Groupings

```python
import pandas as pd
```

```python
# Create three Pandas DataFrames

# This is the basic Carsales DataFrame
Cardata = { "Mercedes": [2, 4, 0, 4, 0, 3], "Ford": [3, 0, 0, 1, 6, 12], "Tata":[9, 3, 4, 1, 0, 0], "Renault":[12, 1, 0, 0, 3, 1]}
Carsales = pd.DataFrame(Cardata)
Carsales.index.rename("Sales place", inplace=True)
# Carsales.rename(index={0: "One", 1: "Two", 2: "Three", 3: "Four", 4: "Five", 5: "Six"}, inplace=True)
Carsales.rename(index={0: "Salesplace_1", 1: "Salesplace_2", 2: "Salesplace_3", 3: "Salesplace_4", 4: "Salesplace_5", 5: "Salesplace_6"}, inplace=True)
Carsales.insert(0, "Sales_place_name", ["Europe 1", "Australia 1", "USA 1", "Asia 1", "Africa 1", "South America 1"], allow_duplicates=False)
Carsales
```

```python
Newsalesplace2 = pd.DataFrame({"Sales_place_name":
                               "South America 1",
                               "Mercedes": 3,
                               "Ford": 2,
                               "Tata": 1,
                               "Renault": 1
                              },
                              index=["Salesplace_7"]
                             )
Newsalesplace3 = pd.DataFrame({"Sales_place_name":
                               "USA 1",
                               "Mercedes": 2,
                               "Ford": 15,
                               "Tata": 1,
                               "Renault": 3
                              },
                              index=["Salesplace_8"]
                             )
Newsalesplace4 = pd.DataFrame({"Sales_place_name":
                               "USA 1",
                               "Mercedes": 1,
                               "Ford": 11,
                               "Tata": 0,
                               "Renault": 0
                              },
                              index=["Salesplace_9"]
                             )
Carsales = pd.concat([Carsales,Newsalesplace2,Newsalesplace3,Newsalesplace4])
Carsales
```

```python
#Sales rating function
def sales_rating(n_sold):
    if n_sold >= 15:
        return "Excellent"
    elif n_sold >=5:
        return "Acceptable"
    else:
        return "Bad"
    
    
```

```python
Carsales2 = Carsales.copy()
Carsales2["Total Sales"] = Carsales[["Mercedes","Ford","Tata","Renault"]].sum(axis=1)
Carsales2
```

```python
Carsales2["Sales Rating"] = Carsales2["Total Sales"].apply(sales_rating)
Carsales2
```

```python
#Create cross tabulation on multiple columns
print(pd.crosstab(Carsales["Sales_place_name"],Carsales["Ford"]))
```

```python
#Create cross tabulation on numerical values vs values
print(pd.crosstab(Carsales["Mercedes"],Carsales["Ford"]))
```

```python
#Create normalized frequency distributions
print(pd.crosstab(Carsales["Mercedes"],Carsales["Ford"],normalize=True))
```

```python
#Create frequency distributions normalized for columns
print(pd.crosstab(Carsales["Mercedes"],Carsales["Ford"],normalize="index"))
```

```python
#Create groups by Sales Places aggregating with mean
print(Carsales.groupby(["Sales_place_name"]).mean())
```

```python
#Create groups by Sales Places aggregating with sum
print(Carsales.groupby(["Sales_place_name"]).sum())
```

```python
print(Carsales.groupby("Sales_place_name", group_keys=True).apply(lambda x:x, include_groups=False))
```

```python
Carsales["Total Sales"] = Carsales[["Mercedes","Ford","Tata","Renault"]].sum(axis=1)
Carsales
```

```python
Carsales["Sales Rating"] = Carsales["Total Sales"].apply(sales_rating)
Carsales
```

```python
print(pd.crosstab(Carsales["Sales_place_name"],Carsales["Sales Rating"]))
```

```python
print(pd.crosstab(Carsales.Sales_place_name,getattr(Carsales, "Sales Rating")))
```

```python
#Use groupby function to create a table for sales rating
print(Carsales.groupby("Sales Rating", group_keys=True).apply(lambda x:x, include_groups=False))
```

```python
print(Carsales.groupby(["Sales Rating", "Sales_place_name"], group_keys=True)["Total Sales"].sum())
```
