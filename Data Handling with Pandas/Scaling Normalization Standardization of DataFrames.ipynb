### Scaling, Normalizaion/ Standardization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading Geyser DataFrame from seaborn library
geyser = sns.load_dataset('geyser')
```

```python
geyser
```

```python
#Create two columns in geyser DataFrame with mean
#corrected data
geyser['mc_duration']=geyser['duration']-geyser['duration'].mean()
geyser['mc_waiting']=geyser['waiting']-geyser['waiting'].mean()
```

```python
geyser
```

```python
#Create two columns with normalized data
geyser['norm_duration']=(geyser['duration']-geyser['duration'].mean())/geyser['duration'].std()
geyser['norm_waiting']=(geyser['waiting']-geyser['waiting'].mean())/geyser['waiting'].std()
```

```python
geyser
```

```python
#Zero One unity normalization or min-max standardization
geyser['Z1_duration']=(geyser['duration']-geyser['duration'].mean())/(geyser['duration'].max()-geyser['duration'].min())

geyser['Z1_waiting']=(geyser['waiting']-geyser['waiting'].mean())/(geyser['waiting'].max()-geyser['waiting'].min())
```

```python
geyser
```

```python
#Plot the features in a 2x2 matirx
fig,((orig,mc),(norm,Z1))=plt.subplots(2,2,constrained_layout=True)
geyser.plot('duration','waiting',kind='scatter',ax=orig)
geyser.plot('mc_duration','mc_waiting',kind='scatter',ax=mc)
geyser.plot('norm_duration','norm_waiting',kind='scatter',ax=norm)
geyser.plot('Z1_duration','Z1_waiting',kind='scatter',ax=Z1)
plt.show()
```
