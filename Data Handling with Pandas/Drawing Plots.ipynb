```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generate data with 500 points
x = np.linspace(0,2*np.pi,500)
#Calculate sine of each x
y = np.sin(x)
```

```python
#Create the plot
plt.figure()
plt.plot(x,y)
plt.title('Sine Curve')
plt.xlabel('$\\theta$')
plt.ylabel('$Sin\\theta$')
plt.grid(False)
plt.show()
```

```python
#Generate data with 500 points
x = np.linspace(-1,1,500)
#Calculate sine of each x
y = np.exp(-x**2)
```

```python
#Create the plot
plt.figure(2)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('$e^{-x^{2}}$')
plt.title('Exponential curve')
```

```python
#Generate data with 500 points
x = np.linspace(0,2*np.pi,500)
#Calculate sine of each x
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
#Create the plot
plt.figure()
plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.title('Sine/Cosine Curve')
plt.xlabel('$\\theta$')
# plt.ylabel('$Sin\\theta$')
plt.grid(False)
plt.legend(['Sin x','Cos x'])
plt.show()
```
