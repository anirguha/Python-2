Write to an excel file with several worksheets

```python
with pd.ExcelWriter('Carsales2.xlsx') as Xlsx_writer:
    for columns in Carsales.columns:
        Carsales[columns].to_excel(Xlsx_writer, sheet_name=columns)
```

Read back the xls file into a DataFrame

```python
all_sheets = pd.read_excel('Carsales2.xlsx',sheet_name=None)
Carsales4 = pd.concat(all_sheets.values(),ignore_index=True)
```

```python
del(Carsales4)
```

```python
file_path = 'Carsales2.xlsx'
all_sheets = pd.read_excel(file_path, sheet_name=None)

Carsales4 = None

for sheet_name, df in all_sheets.items():
    df = df.set_index(df.columns[0])  # Set first column (e.g., 'Index' or 'ID') as index
    # df.columns = [f"{sheet_name}_{col}" for col in df.columns]  
    
    if Carsales4 is None:
        Carsales4 = df
    else:
        Carsales4 = Carsales4.join(df, how='inner')  # Inner join to keep only matching index rows

# Optionally reset index if needed
Carsales4.reset_index(inplace=True)
```

```python
Carsales4.drop(Carsales4.columns[0],axis=1,inplace=True)
```
