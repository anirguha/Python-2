### Pie Charts

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from docutils.nodes import legend
```

```python
#Load dataframe from a seaborne iris library
iris = sns.load_dataset("iris")
```

```python
#Create bins for iris to be used in creating Pie Charts
iris["Sepal_length_class"] = pd.cut(iris.sepal_length, bins=[0,5,6.5,10],labels=["short", "medium", "long"])
iris["Sepal_width_class"] = pd.cut(iris["sepal_width"], bins=[0,2.7,3.4,6],labels=["thin", "normal", "thick"])
```

```python
#Create cross-tabulation between length and width class
iris_1_tab = pd.crosstab(iris["Sepal_length_class"], iris["Sepal_width_class"])
iris_2_tab = iris_1_tab.transpose()

#Add summation columns
iris_1_tab["All Flowers"] = iris_1_tab[iris_1_tab.columns].sum(axis=1)
iris_2_tab["All Flowers"] = iris_2_tab[iris_2_tab.columns].sum(axis=1)

print(iris_1_tab)
print(iris_2_tab)
```

```python
#Create pie charts for table 1
ax = iris_1_tab.plot.pie(y="All Flowers")
ax.set_title("Flower length of 150 sample flowers")
ax.set_ylabel("")
plt.legend(loc="upper left")
plt.showx()
```

```python
#Create pie charts for table 2
ax = iris_2_tab.plot.pie(y="All Flowers")
ax.set_title("Flower width of 150 sample flowers")
ax.set_ylabel("")
plt.legend(loc="upper left")
plt.show()
```

```python
#Create  pie chart fo all widths
fig, ((pie1,pie2),(pie3,pie4)) = plt.subplots(2,2)
explode_dist = (0.1,0,0)
iris_1_tab.plot(y="All Flowers",kind="pie",ax=pie1,autopct="%0.0f%%",
                title="Width classes of 150 sample Flowers", radius=1.3,ylabel="",legend=False, explode=explode_dist)
iris_1_tab.plot(y="thin",kind="pie",ax=pie2,autopct="%0.0f%%",
                title="Thin classes of Flowers", radius=0.8,ylabel="",legend=False)
iris_1_tab.plot(y="thick",kind="pie",ax=pie3,autopct="%0.0f%%",
                title="Thick classes of Flowers", radius=1.0,ylabel="",legend=False)
iris_1_tab.plot(y="normal",kind="pie",ax=pie4,autopct="%0.0f%%",
                title="Normal classes of Flowers", radius=0.5,ylabel="",legend=False)
fig.legend(["short","medium","long"],loc="lower right")
fig.tight_layout()
plt.show()
```

```python
#Create  pie chart fo all lengths
fig, ((pie1,pie2),(pie3,pie4)) = plt.subplots(2,2)
iris_2_tab.plot(y="All Flowers",kind="pie",ax=pie1,autopct="%0.0f%%",
                title="Length classes of 150 sample Flowers", radius=1.3,ylabel="",legend=False)
iris_2_tab.plot(y="short",kind="pie",ax=pie2,autopct="%0.0f%%",
                title="Short classes of Flowers", radius=0.8,ylabel="",legend=False)
iris_2_tab.plot(y="medium",kind="pie",ax=pie3,autopct="%0.0f%%",
                title="Medium classes of Flowers", radius=1.0,ylabel="",legend=False)
iris_2_tab.plot(y="long",kind="pie",ax=pie4,autopct="%0.0f%%",
                title="Long classes of Flowers", radius=0.5,ylabel="",legend=False)
fig.legend(["thin","normal","thick"],loc="lower right")
fig.tight_layout()
plt.show()
```

```python
#Adding shadows to labels
explode_dist = (0.1,0,0)
iris_1_tab.plot(y="All Flowers",kind="pie",autopct="%0.0f%%",
                title="Width classes of 150 sample Flowers", radius=1.0,ylabel="",legend=False, explode=explode_dist)
plt.show()
```

```python
#Chnage the color of the pies in the Pie Chart
sns_colors = sns.color_palette("bright")[0:3]
explode_dist = (0.1,0,0)
iris_1_tab.plot(y="All Flowers",kind="pie",autopct="%0.0f%%",
                title="Width classes of 150 sample Flowers", radius=1.0,ylabel="",legend=False, explode=explode_dist, colors=sns_colors, shadow=True)
plt.legend(loc="upper left")
plt.show()
```
