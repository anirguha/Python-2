### Scatter Plots

```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
iris = sns.load_dataset("iris")
```

```python
#Plot scatter plot
ax = iris.plot(x="sepal_length",y="sepal_width",kind="scatter",color="Black")
ax.set_title("Scatter plot for sepal lengths vs sepal width for 150 sample flowers")
ax.set_xlabel("Sepal Length in cm")
ax.set_ylabel("Sepal Width in cm")
plt.show()
```

```python
#Plot scatter plots with color control and grouped on species feature
fig, ax = plt.subplots()
colors = {"setosa":"Red", "versicolor":"Blue", "virginica": "Yellow"}
for Species, data in iris.groupby("species"):
    data.plot.scatter(x="sepal_length",y="sepal_width",label=Species,
                      color=colors[Species],ax=ax)
ax.set_xlabel("Sepal Length in cm")
ax.set_ylabel("Sepal Width in cm")
ax.set_title("Scatter plot for sepal lengths vs sepal width for 150 sample flowers")
ax.legend(title="Flower Species")
plt.show()
```

```python
#Plot scatter plot for multiple variables with color controls,
#special markers and groups on species feature
fig, ax = plt.subplots()
colors = {"setosa":"Red", "versicolor":"Blue", "virginica": "Yellow"}
for Species, data in iris.groupby("species"):
    data.plot.scatter(x="sepal_length",y="sepal_width",label=Species,
                      color=colors[Species],ax=ax,marker="x")
for Species, data in iris.groupby("species"):
    data.plot.scatter(x="petal_length",y="petal_width",label=Species,
                      color=colors[Species],ax=ax,marker="o")

ax.legend(labels=["sertosa sepal","versicolor sepal","virginica sepal","sertosa petal","versicolor petal",
           "virginica petal"])

ax.set_xlabel("Flower part Length in cm")
ax.set_ylabel("Flower Part Width in cm")
ax.set_title("Scatter plot for flower part lengths vs width for 150 sample flowers")
# ax.legend([sepal_length,petal_length],["Sepals","Petals"])
plt.show()
```

```python
#Plot scatterplot using seaborn
sns.scatterplot(data=iris,x="sepal_length",y="sepal_width",hue="species",
               style="species")
plt.show()
```

```python
#using seaborn's color palettes
sns.scatterplot(data=iris,x="sepal_length",y="sepal_width",hue="species",
               style="species",palette="bright")
plt.show()
```

```python
#Searborn Pairplots
sns.pairplot(data=iris,hue="species")
plt.show()
```

```python
#Plotting multiple variables together
iris["sepal"] = "sepal"
iris["petal"] = "petal"

iris_1 = pd.DataFrame(data=iris[["sepal_length","sepal_width","species","sepal"]])
iris_1.columns = ["length","width","species","measure"]
iris_2 = pd.DataFrame(data=iris[["petal_length","petal_width","species","petal"]])
iris_2.columns = ["length","width","species","measure"]
iris2 = pd.concat([iris_1,iris_2])
iris2
```

```python
#plot seaborn scatter plot over 
sns.scatterplot(data=iris2,x="width",y="length",hue="species",style="measure")
plt.show()
```
