### Array/ Matrix visualization

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randn(5,5)
```

```python
data.ndim, data.size, data.shape
```

```python
fig, ax = plt.subplots()

#Set handle for the image
im = ax.imshow(data, cmap="viridis", aspect="auto",
               interpolation="nearest", origin="lower")
plt.colorbar(im, ax=ax)
ax.set_title("Matrix Visualization")
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)

#Set handle for the image
im = ax1.imshow(data, cmap="magma",
               interpolation="nearest", origin="lower")
ax1.set_aspect("equal")
plt.colorbar(im, ax=ax1)
ax1.set_title("Matrix Visualization")

im = ax2.imshow(data.transpose(), cmap="magma", 
               interpolation="nearest", origin="lower")
ax1.set_aspect("equal")
plt.colorbar(im, ax=ax2)
ax2.set_title("Transposed Matrix Visualization")

plt.show()
```

```python
#Create a symmetric array
data_symm = data * data.transpose()
```

```python
data_symm
```

```python
fig, ax = plt.subplots()

#Set handle for the image
im = ax.imshow(data_symm, cmap="viridis", aspect="auto",
               interpolation="nearest", origin="lower")
plt.colorbar(im, ax=ax)
ax.set_title("Matrix Visualization")
plt.show()
```

```python
data == data_symm.transpose()
```

```python
data - data_symm.transpose()
```

```python
data_symm = (data + data.transpose()) / 2
```

```python
plt.imshow(data_symm, interpolation="nearest",origin="lower")
plt.show()
```

```python
data_symm == data_symm.transpose()
```

```python
data_symm - data_symm.transpose()
```

```python
data_symm_inv = np.linalg.inv(data_symm)
```

```python
data_symm_inv
```

```python
data_id = np.matmul(data_symm, data_symm_inv)
```

```python
fig, (ax1,ax2, ax3) = plt.subplots(1,3)

#Set handle for the image
im = ax1.imshow(data_symm, cmap="magma",
               interpolation="nearest", origin="lower")
ax1.set_aspect("equal")
plt.colorbar(im, ax=ax1,shrink=0.25)
ax1.set_title("Original Matrix",fontsize=10)

im = ax2.imshow(data_symm_inv.transpose(), cmap="magma", 
               interpolation="nearest", origin="lower")
ax2.set_aspect("equal")
plt.colorbar(im, ax=ax2, shrink=0.25)
ax2.set_title("Inverted Matrix",fontsize=10)

im = ax3.imshow(data_id, cmap="magma", 
               interpolation="nearest", origin="lower")
ax3.set_aspect("equal")
plt.colorbar(im, ax=ax3,shrink=0.25)

ax3.set_title("Matrix Product",fontsize=10)

plt.tight_layout()
plt.show()
```

```python
##Eigen decomposition of the matrix
data_symm2 = (data + data.T)/2
eigenvalues, eigenvectors = np.linalg.eigh(data_symm2)

list_eig = [i for i in zip(eigenvalues,eigenvectors)]
```

```python
for idx, values in enumerate(list_eig):
    print('Eigen Values:',values[0])
    print('Eigen Vectors:',values[1])
```

```python
eig_diag = np.diag(eigenvalues)
eig_diag
```

```python
#Reconstruct the matrix from eigenvector matrices
eigenvectors_inv = np.linalg.inv(eigenvectors)
data_recon = eigenvectors @ eig_diag @ eigenvectors_inv
```

```python
data_recon
```

```python
np.allclose(data_symm2, data_recon)
```

```python
data_recon - data_symm2
```

```python
fig, ((ax1,ax2, ax3), (ax4, ax5,ax6)) = plt.subplots(2,3)

#Set handle for the image
im = ax1.imshow(eigenvectors, cmap="magma",
               interpolation="nearest", origin="lower")
ax1.set_aspect("equal")
plt.colorbar(im, ax=ax1,shrink=0.25)
ax1.set_title("Eigenvectors",fontsize=10)

im = ax2.imshow(eig_diag, cmap="magma", 
               interpolation="nearest", origin="lower")
ax2.set_aspect("equal")
plt.colorbar(im, ax=ax2, shrink=0.25)
ax2.set_title("Diagonal matrix",fontsize=10)

im = ax3.imshow(eigenvectors_inv, cmap="magma", 
               interpolation="nearest", origin="lower")
ax3.set_aspect("equal")
plt.colorbar(im, ax=ax3,shrink=0.25)

ax3.set_title("Inverse of eigenvector matrix",fontsize=10)

im = ax4.imshow(data_symm2, cmap="magma", 
               interpolation="nearest", origin="lower")
ax4.set_aspect("equal")
plt.colorbar(im, ax=ax4,shrink=0.25)

ax4.set_title("Original Matrix",fontsize=10)

im = ax5.imshow(data_recon, cmap="magma", 
               interpolation="nearest", origin="lower")
ax5.set_aspect("equal")
plt.colorbar(im, ax=ax5,shrink=0.25)

ax5.set_title("Reconstructed Matrix",fontsize=10)

im = ax6.imshow(np.isclose(data_symm2, data_recon), cmap="magma", 
               interpolation="nearest", origin="lower")
ax6.set_aspect("equal")
plt.colorbar(im, ax=ax6,shrink=0.25)

ax6.set_title("Reconstructed Matrix",fontsize=10)

plt.tight_layout()
plt.show()
```
