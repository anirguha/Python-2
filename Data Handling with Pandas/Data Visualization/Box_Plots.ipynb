```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load seaborn dataset iris
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
#Plot Box Plot
ax = iris.boxplot(column="sepal_length",showmeans=True,meanline=True,
                  grid=False)
plt.suptitle("")
ax.set_title("Box Plot for Sepal Lengths of 150 flowers",fontsize=16)
plt.show()
```

```python
#Plot BoxPlot for a single column separated by species of flower
ax = iris.boxplot(column="sepal_length",by="species",showmeans=True,
                  meanline=True,grid=False)
plt.suptitle("")
ax.set_title("Box Plot for Sepal Lengths of 150 flowers separated by species",fontsize=16)
plt.show()
```

```python
#seaborn boxplot
from matplotlib.patches import Patch

# Set palette
palette = sns.color_palette("Set2", 3)
species_list = iris["species"].unique()

# Plot without hue (colors still apply)

sns.boxplot(x="species",y="sepal_length",data=iris,hue="species",legend=False,
            dodge=False,width=0.3, palette=palette)
plt.suptitle("")
plt.title("Box Plot for Sepal Lengths of 150 flowers separated by species",fontsize=16)

# Manually create legend
handles = [Patch(color=palette[i], label=species_list[i]) for i in range(len(species_list))]
plt.legend(handles=handles, title="Species", loc="upper left")

plt.suptitle("")
plt.title("Box Plot for Sepal Lengths of 150 flowers separated by species",fontsize=16)
plt.show()
```

```python
sns.boxenplot(data=iris,x="species",y="sepal_length",hue="species",width=0.4,dodge=False)
plt.show()
```

```python
#seaborn violin plot based on the species feature
sns.violinplot(data=iris,y="sepal_length",x="species",hue="species",dodge=False,width=0.3)
plt.show()
```

```python
#seaborn warm or bee swarm plots
sns.swarmplot(data=iris,x="species",y="sepal_length",hue="species")
plt.show()
```

```python
#combining swarmplot with violinplot
sns.catplot(data=iris,x="species",y="sepal_length",kind="boxen",hue="species",dodge=False,width=0.5)
sns.swarmplot(data=iris,x="species",y="sepal_length",hue="species")
plt.show()
```
