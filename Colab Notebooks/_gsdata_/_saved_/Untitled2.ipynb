```python
import time

for i in range(1, 11):
    print(f"Processing item {i}/10", end="\r",flush=True)
    time.sleep(0.5)
```

```python
import sys, time

for i in range(1, 11):
    sys.stdout.write(f"\rProcessing item {i}/10")
    sys.stdout.flush()
    # time.sleep(0.5)
```

```python
from tqdm import tqdm
import time

for i in tqdm(range(10), desc="Processing"):
    time.sleep(0.5)
```
