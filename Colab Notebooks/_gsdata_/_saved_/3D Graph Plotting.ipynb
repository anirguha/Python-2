```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(1)
ax = plt.axes(projection='3d')
ax.set_box_aspect([1,1,1])

z = np.linspace(0,1,500)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

ax.plot3D(x,y,z,'blue')
ax.set_title('3D Line Plot')

plt.show()
```
