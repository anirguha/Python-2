```python
import torch
from transformers import AutoModelForCausalLM, GPT2Tokenizer

!pip install torchinfo
from torchinfo import summary
```

```python
gpt2 = AutoModelForCausalLM.from_pretrained("gpt2")
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
```

```python
gpt2
```

```python
gpt2.config
```

```python
summary(gpt2, input_data=torch.randint(0, gpt2.config.vocab_size, (1, 1024), dtype=torch.long), col_names=["input_size", "output_size", "num_params", "trainable"], row_settings=["var_names"])
```

```python
x = 'Hello how are you today?'

input_ids = tokenizer.encode(x, return_tensors='pt')

out = gpt2.generate(input_ids, max_length=50, do_sample=True, temperature=1., pad_token_id=tokenizer.eos_token_id)

print(tokenizer.decode(out[0].tolist()))
```
