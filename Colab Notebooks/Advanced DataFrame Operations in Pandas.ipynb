# Practice Exercise

```python
# @title
#Find out the players that meet the below criteria:
#they're of English nationality
#their market value is more than twice the average value in the league
#they have more than 4,000 views or a new signing but not both
```

```python
# @title
#Playes of English natioanlity
players_english = players.nationality == 'England'
players.loc[players_english]
```

```python
# @title
#condition to extract players whose market values are more than twice the avarage market value
average_market_value = players.market_value.mean()
players_twice_average = players.market_value > (average_market_value * 2)
players.loc[players_twice_average]
```

```python
# @title
#Exrtacting playes who have more than 4,000 views or a new signing but not both
players_morethan_4000_views = players.page_views > 4000

players_new_signing = players.new_signing == 1
players_new_signing_xor_4000_views = players_morethan_4000_views ^ players_new_signing
players.loc[players_new_signing_xor_4000_views]
```

```python
# @title
#combining all conditions
new_players_df = players.loc[players_english & players_twice_average & players_new_signing_xor_4000_views]
new_players_df
```

### **Skill Challenge # 3**

```python
#Create a function that accespts a single parameter x
#if x is less than 200 returns a string 'relatively unknown'
#if x is between 200 and 600(non-inclusize) returns 'kind of popular'
#if x is between 600 and 2000 (non-inclusive) returns 'popular'
# otherwise return 'super popular'
```

```python
def popularity_class(x):
  if x < 200:
    return 'relatively unknown'
  elif 200 <= x < 600:
    return 'kind of popular'
  elif 600 <= x < 2000:
    return 'popular'
  else:
    return 'super popular'
```

```python
#apply function to players page_views column
df = players.page_views.transform(popularity_class)
df
```

```python
#Add the output from the step above as a new column in playes DF.
players['popularity_class'] = df
players.head()
```

```python
# count the number of super popular players
sum(players.popularity_class == 'super popular')
players.popularity_class.value_counts()
```
