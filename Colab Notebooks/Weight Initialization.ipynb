```python
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
# Create weight matrices
weights = {
    'norm': nn.Linear(50,50),
    'xavier': nn.Linear(50,50),
    'kaiming': nn.Linear(50,50)

}
```

```python
# Plot default weight distributions
plt.figure(figsize=(10,10))

for name, w in weights.items():

  # Get the distribution
  y, x = np.histogram(w.weight.detach().flatten(), bins=40)
  plt.plot(x[:-1], y, label=name)
  plt.gca().set(xlabel='Weight Values', ylabel='Count', title='Distribution of weights before initialization')

plt.legend()
plt.show()
```

```python
# Initialize the matrices (weights and bias)
torch.nn.init.normal_(weights['norm'].weight, mean=0, std=3)
torch.nn.init.xavier_normal_(weights['xavier'].weight)
torch.nn.init.kaiming_normal_(weights['kaiming'].weight);
```

```python
# Plot default weight distributions
plt.figure(figsize=(10,10))

for name, w in weights.items():

  # Get the distribution
  y, x = np.histogram(w.weight.detach().flatten(), bins=40)
  plt.plot(x[:-1], y, label=name)
  plt.gca().set(xlabel='Weight Values', ylabel='Count', title='Distribution of weights after initialization')

plt.legend()
plt.show()
```
