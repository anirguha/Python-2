```python
import torch
import torch.nn as nn
import torch.nn.functional as F

import numpy as np
```

# Generate random data

```python
nObservations = 5
nFeatures = 10

data = np.random.randn(nObservations, nFeatures)
```

# Define a FFN model

```python
class FFNmodel(nn.Module):
  def __init__(self):
    super().__init__()
    self.fc1 = nn.Linear(nFeatures, 32)
    self.fc2 = nn.Linear(32, 64)
    self.fc3 = nn.Linear(64, 1)

  def forward(self,x):
    print('Shape of input for layer1',x.shape)
    x = F.relu(self.fc1(x))
    print('Shape of output of layers1/ input for layer2',x.shape)
    x = F.relu(self.fc2(x))
    print('Shape of output of layers2/ input for layer3',x.shape)
    x = self.fc3(x)
    print('Shape of output of layers3',x.shape)
    return x
```

```python
def createFFNmodel():
  model = FFNmodel()

  return model
```
