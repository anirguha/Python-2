```python
import torch
from transformers import AutoModelForCausalLM, GPT2Tokenizer

import numpy as np
import matplotlib.pyplot as plt
```

```python
tokenizer = AutoModelForCausalLM.from_pretrained('gpt2')
```

```python
model_ids = {
    'small': 'gpt2',
    'medium': 'gpt2-medium',
    'large': 'gpt2-large',
    'xl': 'gpt2-xl'
}

# Load all models in a dictionary
models = {name: AutoModelForCausalLM.from_pretrained(model_id) for name, model_id in model_ids.items()}
```

```python
for name, model in models.items():
  print(f'{sum(p.numel() for p in model.parameters()):13,} parameters in gpt2-{name} model')
```
