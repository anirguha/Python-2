```python
import pandas as pd
from string import ascii_uppercase, ascii_lowercase
```

```python
alphabet = pd.Series(list(ascii_lowercase),index=map(lambda x: 'label_' + x, ascii_uppercase),name='letter')
alphabet
```

Exercise

```python

def every_fifth(x):
  return [True if (i+1)%5==0 else False for i in range(x.size)]

alphabet.loc[every_fifth]
```

Using a generator

```python

def every_fifth(x):
  for i in range(x.size):
    if (i+1)%5==0:
      yield True
    else:
      yield False

alphabet.loc[every_fifth]
```
