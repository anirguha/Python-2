# Munging Data

```python
strip_nutrition = nutrition.astype(str).replace('[^a-zA-Z]','',regex=True)
```

```python
strip_nutrition.head()
```

```python
units = strip_nutrition.mode()
units
```

```python
# units = units.replace('',np.nan)
# units = units.infer_objects(copy=False)
# units.dropna(how='all',axis=1)

units = units.loc[:,units.iloc[0] != ""]
units


# for i in units:
#     print(i,units[i].iloc[0])
```

```python
#Create a mapper like <columnname: columnname_unit
mapper = {i:i + '_' + units[i].iat[0] for i in units}
```

```python
mapper
```

```python
nutrition.rename(columns=mapper,inplace=True)
```

```python
nutrition
```

```python
nutrition.replace('[a-zA-Z]','',regex=True,inplace=True)
```

# Skill Challenge 2

```python
 #Exclude food items that contain at least one NaN
df = nutrition.dropna(how='any',axis=0)
df.shape[0]
```

```python
#Extract food items that have between 20 and 40 mg of Vitamin C and find the one with minimum calories
df[df.vitamin_c_mg.between(20,40)].calories.nsmallest(1)
```

```python
#Find the number of food items that are within 2-3 standard deviations from the mean
z_score = ((df.vitamin_c_mg - df.vitamin_c_mg.mean()).abs())/df.vitamin_c_mg.std()
df[z_score.between(2,3)].shape[0]
```
