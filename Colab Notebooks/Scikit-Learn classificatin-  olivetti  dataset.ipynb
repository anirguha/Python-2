```python
import matplotlib.pyplot as plt

val = 30

fig, ax = plt.subplots(2, 5, figsize=(12, 6))
images_to_plot = olivetti_faces.images[olivetti_faces.target == val][:10] # Select the first 10 images from class 1
targets_to_plot = olivetti_faces.target[olivetti_faces.target == val][:10] # Select the first 10 targets from class 1

for i, axi in enumerate(ax.flat):
    axi.imshow(images_to_plot[i], cmap='gray')
    axi.set(xticks=[], yticks=[],
    xlabel=targets_to_plot[i]) # Use the corresponding target
plt.show()
```
