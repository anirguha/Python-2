```python
#Split the Official Time into three separate columns. Store the dataframe in a varaible called final_times.
```

```python
final_times = boston['Official Time'].str.split(':',expand=True)
```

```python
final_times.head()
```

```python
#In the final_times dataframe create a new column called total that contains the total in seconds. Add the column to the boston dataframe
```

```python
final_times = final_times.astype(int)
```

```python
final_times
```

```python
final_times['total'] = final_times.apply(lambda x: x.iat[0]*60*60 + x.iat[1]*60 + x.iat[2],axis=1)
```

```python
final_times.total.notnull().value_counts()
```

```python
#Do the runners in boston dataframe who ran in 2016 has a higher or lower average time?
```
