```python
class Money:
  def __init__(self,x):
    self.x = x
  def __add__(self,other):
    if isinstance(other,Money):
      return Money(self.x + other.x)
    else:
      # Handle cases where 'other' is not a Money object (e.g., an integer)
      return Money(self.x + other)
  def __str__(self):
    return str(self.x)
```

```python
m1 = Money(5)
m2 = Money(10)

print(m1+3)

print(m1+m2)
```
