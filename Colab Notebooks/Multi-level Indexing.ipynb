# **Creating MultiIndex Columns in Pandas**

```python
#Using pd.MultiIndex.from_tuples() method

import pandas as pd
import numpy as np

# Define multi-level column headers
columns = pd.MultiIndex.from_tuples([
    ('Math', 'Algebra'),
    ('Math', 'Geometry'),
    ('Science', 'Physics'),
    ('Science', 'Chemistry')
])

# Create data
data = np.random.randint(50, 100, size=(3, 4))

# Create DataFrame
df = pd.DataFrame(data, columns=columns, index=['Student1', 'Student2', 'Student3'])

print(df)
```

```python
#Using nested list pd.MultiIndex.from_product() method

# Define levels
subjects = ['Math', 'Science']
topics = ['Algebra', 'Geometry']

#Create product of levels
columns = pd.MultiIndex.from_product([subjects,topics])

#Create DataFrame
df = pd.DataFrame(data=np.random.randint(0,high=100,size=(3,4)),columns=columns,index=['Student1','Student2','Student3'])
print(df)
```

```python
#Setting columns after dataframe creation
df = pd.DataFrame(
    [[90, 85, 80, 75], [88, 82, 78, 72]],
    columns=[('Math', 'Algebra'), ('Math', 'Geometry'),
             ('Science', 'Physics'), ('Science', 'Chemistry')],
    index = ['Student1','Student2']
)

df.columns = pd.MultiIndex.from_tuples(df.columns)

df
```

```python
#Accessing data
df.loc['Student1',('Math','Algebra')]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Needed for 3D plotting

# Generate a 3D array of random integers
arr = np.random.randint(0, 10, size=(30, 3))  # Shape: (30 points, 3 coordinates)

# Extract x, y, z coordinates
x = arr[:, 0]
y = arr[:, 1]
z = arr[:, 2]

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, c='blue', marker='o')

# Set labels
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Scatter Plot')

plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import numpy as np

# Create a 3D array of shape (2, 3, 4)
arr = np.random.randint(0, 10, size=(2, 3, 4))

# Get the shape
depth, rows, cols = arr.shape

# Get the grid of indices
x, y, z = np.meshgrid(np.arange(cols), np.arange(rows), np.arange(depth))

# Flatten all arrays
x = x.flatten()
y = y.flatten()
z = z.flatten()
values = arr.flatten()

# 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Use values to determine color or size
sc = ax.scatter(x, y, z, c=values, cmap='viridis', s=60)

# Labels
ax.set_xlabel('Column Index')
ax.set_ylabel('Row Index')
ax.set_zlabel('Depth Index')
ax.set_title('3D Scatter from 3D Array')

# Add color bar
plt.colorbar(sc)
plt.show()
```
