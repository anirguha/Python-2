```python
import os
import requests
from pathlib import Path

def download_github_folder(user, repo, branch, folder_path, save_dir="."):
    """
    Recursively download a folder from a public GitHub repository.

    Args:
        user (str): GitHub username or org (e.g. "anirguha")
        repo (str): Repository name (e.g. "Python-2")
        branch (str): Branch name (e.g. "main")
        folder_path (str): Path inside the repo (e.g. "Important PyTorch Modules")
        save_dir (str): Local directory to save files to (default: current dir)
    """

    api_url = f"https://api.github.com/repos/{user}/{repo}/contents/{folder_path}?ref={branch}"
    response = requests.get(api_url)
    response.raise_for_status()

    items = response.json()
    Path(save_dir).mkdir(parents=True, exist_ok=True)

    for item in items:
        item_path = Path(save_dir) / item["name"]

        if item["type"] == "file":
            download_url = item["download_url"]
            print(f"⬇️  Downloading file: {item_path}")
            r = requests.get(download_url)
            r.raise_for_status()
            with open(item_path, "wb") as f:
                f.write(r.content)

        elif item["type"] == "dir":
            print(f"📁 Entering folder: {item['path']}")
            download_github_folder(
                user, repo, branch, item["path"], save_dir=item_path
            )

    print(f"✅ Finished downloading folder: {folder_path}")
```

```python
download_github_folder(
    user="anirguha",
    repo="Python-2",
    branch="master",
    folder_path="Helper Functions/Important PyTorch Modules",  # folder in repo
    save_dir="./downloaded_modules"           # local directory
)
```

```python
import requests
from pathlib import Path

def download_from_github_raw(user, repo, branch, filepath, save_dir="."):
    """Download a raw file from a GitHub repo"""
    raw_url = f"https://raw.githubusercontent.com/{user}/{repo}/{branch}/{filepath}"
    save_path = Path(save_dir) / Path(filepath).name
    response = requests.get(raw_url)
    response.raise_for_status()
    save_path.write_bytes(response.content)
    print(f"✅ Saved {save_path}")
    return save_path
```
