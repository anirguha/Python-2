```python
import numpy as np
from scipy.linalg import eig

# Example matrices
A = np.array([[4, 2],
              [1, 3]], dtype=float)

B = np.array([[1, 0],
              [0, 2]], dtype=float)

# Solve the generalized eigenvalue problem A v = lambda B v
eigenvalues, eigenvectors = eig(A, B)

print("Generalized eigenvalues:")
print(eigenvalues)

print("\nGeneralized eigenvectors (columns):")
print(eigenvectors)

# Verify: A v ≈ lambda B v
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]
    lhs = A @ v
    rhs = lam * (B @ v)
    print(f"\nEigenpair {i}:")
    print("A @ v =", lhs)
    print("lambda * B @ v =", rhs)
    print("Close?", np.allclose(lhs, rhs))
```
