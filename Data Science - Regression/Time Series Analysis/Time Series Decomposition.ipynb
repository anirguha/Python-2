## Time Series Decomposition

```python
import pandas as pd
```

```python
hourly_weather = pd.read_csv(
    "../Data/madrid_weather.csv",
    usecols=['time','temperature'],
    index_col='time',
    parse_dates=['time'],
    # date_format="%Y-%m-%d %H:%M:%S"
)
hourly_weather.index.dtype
```

```python
# import matplotlib.pyplot as plt
# plt.plot(hourly_weather.iloc[1:168,:])
# plt.show()

import matplotlib.pyplot as plt
plt.plot(hourly_weather)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(hourly_weather[:168])

result.plot();
```

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(hourly_weather);
```

```python
monthly_weather = pd.read_csv(
    "../Data/madrid_weather.csv",
    usecols=['time','temperature'],
    index_col='time',
    parse_dates=['time'],
    # date_format="%Y-%m-%d %H:%M:%S"
).resample("ME").mean()

monthly_weather
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

seasonal_decompose(monthly_weather).plot();
```

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(monthly_weather);
```
