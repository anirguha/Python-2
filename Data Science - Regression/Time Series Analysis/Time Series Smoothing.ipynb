## Time Series Smoothing

```python
import pandas as pd
```

```python
electricity_df = pd.read_csv(
    "../Data/powerconsumption.csv",
    usecols=['Datetime','PowerConsumption_Zone1'],
    index_col=['Datetime'],
    parse_dates=['Datetime']
)
electricity_df = electricity_df.resample('h').mean()
electricity_df.head()
```

```python
electricity_df.plot();
```

```python
# Calculate 1 week rolling average
electricity_df.rolling(window=24*7).mean().plot();
```

```python
# Calculate 30 day rolling average
electricity_df.rolling(window=24*30).mean().plot();
```

```python
import matplotlib.pyplot as plt
plt.plot(electricity_df)
plt.plot(electricity_df.rolling(window=24*7).mean(),linewidth=3)
plt.plot(electricity_df.rolling(window=24*30).mean(),linewidth=3)
plt.legend(['Hourly consumtion','Rolling Weekly','Rolling Monthly'])
plt.show();
```
