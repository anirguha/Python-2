## **Cosine Similarity**

```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_excel('../Data/Movie_Ratings.xlsx',0)
```

```python
movies.head()
```

```python
movies_df = movies.drop(['Movie_ID'], axis=1)
```

```python
movies_df.info()
```

```python
movies_df = movies_df.set_index(['Movie_Title'])
movies_df.index.name = None
movies_df
```

```python
movies_df.loc[['Toy Story (1995)','Get Shorty (1995)']]
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity(movies_df.loc[['Toy Story (1995)','Get Shorty (1995)']])[0,1]
```

```python
cosine_matrix = cosine_similarity(movies_df)
```

```python
cosine_df = pd.DataFrame(cosine_matrix, index=movies_df.index, columns=movies_df.index)
```

```python
cosine_df.shape
```

```python
## Costine similarity of Tyoy Story with other movies
cosine_df.loc[:,['Toy Story (1995)']].sort_values('Toy Story (1995)',ascending=False).head(6)
```
