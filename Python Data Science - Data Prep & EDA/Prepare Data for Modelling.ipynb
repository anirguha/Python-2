### Feature Engineering

```python
## Create feature based on the school going children of the customer
customers.info()
```

```python
customers.head()
```

```python
has_kids = (pd.get_dummies(customers[['Customer ID','Has School Aged Children']],drop_first=True,dtype=int)
                .rename(columns={'Has School Aged Children_Yes': 'Has Kids'})
 )
```

```python
has_kids.head()
```

```python
model_df
```

```python
## Merge the dataframe to the model dataframe
model_df = model_df.merge(has_kids,on='Customer ID',how='left')
```

```python
model_df
```

```python
## Make Purchase location numeric
df_april_may.head()
```

```python
location_dummy = pd.get_dummies(df_april_may["Purchase Location"],drop_first=True,dtype='int')
```

```python
## Concatenate the dummy location to the April, May sales table by columns(horizontally)
location_df = pd.concat([df_april_may,location_dummy],axis=1)[['Customer ID','Online']]
```

```python
location_df
```

```python
## Caluclate the percentage of online purchases by customer
in_person_pct = (
                    (
                            1 -
                        location_df.groupby('Customer ID')['Online'].sum()/
                        location_df.groupby('Customer ID')['Online'].count()
                             ) * 100
                    ).rename('In_person pct').reset_index()
```

```python
in_person_pct
```

```python
## Add to the model dataframe
model_df = pd.concat([model_df,in_person_pct],axis=1)
```

```python
# y/ target = 'June Purchases'
# x/ feature = rest of the columns
# Drop Customer ID column as it has no value
model_df = model_df.loc[:,['June Purchases','Total Spend','Has Kids','Adults','Children','Teens','In_person pct']]
```
