```python
import torch
import torch.nn.functional as F
import torch.nn as nn

import matplotlib.pyplot as plt
```

```python
# Example 1: Resizing a 2D image using bilinear interpolation
input_tensor = torch.randn(1, 3, 10, 10) # Batch, Channels, Height, Width
print(f"Input tensor size (by size): {input_tensor.shape}")
```

```python
output_tensor_size = F.interpolate(input_tensor, size=(20, 20), mode='bilinear', align_corners=False)
print(f"Output tensor size (by size): {output_tensor_size.shape}")
```

```python
# Example 2: Upsampling a 2D image by a scale factor
output_tensor_scale = F.interpolate(input_tensor, scale_factor=2, mode='nearest')
print(f"Output tensor size (by scale_factor): {output_tensor_scale.shape}")
```

```python
# Example 3: Resizing a 3D volume using trilinear interpolation
input_volume = torch.randn(1, 1, 5, 5, 5) # Batch, Channels, Depth, Height, Width
output_volume = F.interpolate(input_volume, size=(10, 10, 10), mode='trilinear', align_corners=False)
print(f"Output volume size: {output_volume.shape}")
```

```python
plt.imshow(input_tensor[0,0,:,:])
plt.colorbar()
plt.show();
```

```python
fig, ax = plt.subplots(1,2,figsize=(20,8))
img1 = ax[0].imshow(output_tensor_scale[0,0,:,:])
plt.colorbar(img1)

img2 = ax[1].imshow(output_tensor_size[0,0,:,:])
plt.colorbar(img2)

plt.show();
```
