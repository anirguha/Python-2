### Check Python version in Driver node

```python
from pyspark.sql import SparkSession

python_path = "/opt/anaconda3/envs/pyspark_env/bin/python"

spark = (
    SparkSession.builder
    .appName("ForcePythonFix")
    .config("spark.pyspark.python", python_path)
    .config("spark.pyspark.driver.python", python_path)
    .getOrCreate()
)
```

```python
import sys
print("Driver Python version:", sys.version)
```

### Check Python version in worker node

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

rdd = spark.sparkContext.parallelize([0])
worker_python_version = rdd.map(lambda x: __import__('sys').version).collect()

print("Worker Python version:", worker_python_version[0])
```
